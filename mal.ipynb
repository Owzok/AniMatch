```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import re
```

```python
keys = ['Japanese', 'English', 'Type', 'Episodes', 'Status', 'Aired', 'Premiered', 'Broadcast', 'Producers', 'Licensors', 'Studios', 'Source', 'Genres', 'Themes', 'Duration', 'Rating', 'Score', 'Ranked', 'Popularity', 'Members', 'Favorites']
```

```python
URL = "https://myanimelist.net/anime/"
r = requests.get(URL+str(200))
soup = BeautifulSoup(r.content)
soup
```

```python
class MalScrapper:
    def __init__(self):
       self.URL = "https://myanimelist.net/anime/"

    def generate_data(self, index):
        r = requests.get(self.URL+str(index))
        soup = BeautifulSoup(r.content)

        anime_info = {key: '' for key in keys}

        try:
            anime_info["Title"] = soup.find("h1", class_="title-name").text.strip()
        except Exception as e:
            print(f"Title error on {index}:\n",e)
            return

        anime_info["Id"] = index

        try:
            anime_info["Positive_reviews"] = soup.find("div", class_="recommended").find("strong").text
            anime_info["Negative_reviews"] = soup.find("div", class_="not-recommended").find("strong").text
        except Exception as e:
            print(f"Reviews error on {index}:\n",e)
        
        for i in soup.find_all("div", class_="spaceit_pad"):
            category = i.text.strip().split(":")
            if len(category) > 1:
                anime_info[category[0]] = category[1].replace("\n", "")
        return anime_info

    def query(self, start, end):
        data = []
        for i in range(start, end):
            x = self.generate_data(i)
            if x != None:
                print(i)
                data.append(x)
        df = pd.DataFrame(data)
        df.to_csv(f"data_{start}_{end}.csv",index=False)
        return
```

```python
import time
```

```python
mal = MalScrapper()
```

```python
start = 1
end = 60000
while True:
    if start >= end:
        break
    mal.query(start, start + 100)
    time.sleep(330)  # 330 seconds = 5.5 minutes
    start += 100 
```
