```python
import pandas as pd
```

```python
main = pd.read_csv("data_anime_clean.csv")
df = pd.read_csv("anime_with_synopsis.csv")

main.head(3), df.head(3)
```

```python
new_df = main[["Id", "Type", "Episodes", "Studios", "Members"]]
new_df
```

```python
df.head(1)
```

```python
new_df.head(1)
```

```python
final_df = new_df.merge(df, how="inner", left_on="Id", right_on="MAL_ID")
```

```python
final_df["Studios"] = final_df["Studios"].apply(lambda x: x.split(",")[0])
final_df.drop(["MAL_ID"], axis=1, inplace=True)
```

```python
final_df = final_df[final_df["Members"] > 10000]
```

```python
final_df.iloc[1000:]
```

```python
final_df.to_csv("final.csv", index=False)
```
