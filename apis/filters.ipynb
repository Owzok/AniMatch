```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./data/data_anime_clean.csv")
df
```

```python
df = df.replace("Unknown", np.nan)
df = df[df.Aired_Start.isna() == False]
df
```

```python
def get_year(x):
    try:
        res = x.split(",")[1][1:]
        return res
    except:
        if len(x) == 4:
            return x
        else:
            return x.split(" ")[1]
```

```python
df['Year'] = df['Aired_Start'].apply(get_year)
df.head(1)
```

```python
df.columns
```

```python
df = df[df.Members > 4000]
```

```python
df = df.drop(["Name", "Genres", "English", "Type", "Producers", "Licensors", "Studios", "Source", "Duration", "Rating", "Ranked", "Popularity", "Members", "Favorites", "Aired_Start", "Aired_End", "Premiered_Year", "Premiered_Season"], axis=1)
```

```python
df['Episodes'] = df['Episodes'].fillna(1000)
```

```python
df.to_csv("./data/data4filter.csv", index=False)
```
