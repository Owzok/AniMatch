```python
word = "Themes, ParodyParody,         SchoolSchool"
```

```python
def removeDupes(x):
    final = []
    lst = x.split(",")
    for i in lst:
        y = i.strip()
        z = int(len(y)/2)

        if y[:z] == y[z:]:
            final.append(y[z:])
        else:
            final.append(y)
    return ','.join(final)
```

```python
removeDupes(word)
```

```python
df.Licensors = df.Licensors.apply(lambda x: str(x).strip().replace("\n", "").replace("\t", ""))
df.Licensors = df.Licensors.apply(removeDupes)
df.Licensors
```

```python
df.Studios = df.Studios.apply(lambda x: str(x).strip().replace("\n", "").replace("\t", ""))
df.Studios = df.Studios.apply(removeDupes)
df.Studios
```

```python
df.Source = df.Source.apply(lambda x: str(x).strip().replace("\n", "").replace("\t", ""))
df.Source = df.Source.apply(removeDupes)
df.Source
```

```python
df.Genres = df.Genres.apply(lambda x: str(x).strip().replace("\n", "").replace("\t", ""))
df.Genres = df.Genres.apply(removeDupes)
df.Genres
```

```python
df.Related_content = df.Related_content.apply(lambda x: str(x).strip().replace("\n", "").replace("\t", ""))
df.Related_content = df.Related_content.apply(removeDupes)
df.Related_content
```

```python
def find_prequel(w):
    for i in w.split(":"):
        if i == 'Prequel':
            return 1
    return 0

def find_hentai(w):
    for i in w.split(" "):
        if i == 'Hentai':
            return 1
    return 0
```

```python
df['Prequel'] = df['Related_content'].apply(find_prequel)
df['Prequel']
```

```python
df['Hentai'] = df['Genres'].apply(find_hentai)
df[df['Hentai'] == 1]['Title']
```
