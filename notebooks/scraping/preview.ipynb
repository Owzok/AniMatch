```python
def removeDupes(x):
    final = []
    lst = x.split(",")
    for i in lst:
        y = i.strip()
        z = int(len(y)/2)

        if y[:z] == y[z:]:
            final.append(y[z:])
        else:
            final.append(y)
    return ','.join(final)
```

```python
removeDupes(word)
```

```python
df.Licensors = df.Licensors.apply(lambda x: str(x).strip().replace("\n", "").replace("\t", ""))
df.Licensors = df.Licensors.apply(removeDupes)
df.Licensors
```

```python
def find_prequel(w):
    for j in w.split(","):
        for i in j.split(":"):
            if i == 'Prequel':
                return 1
    return 0

def find_hentai(w):
    for i in w.split(" "):
        if i == 'Hentai':
            return 1
    return 0
```
