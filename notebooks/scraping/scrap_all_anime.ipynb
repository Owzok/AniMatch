### Function to scrap all that given a link

````python
def scrap_all_raw(link):
    show_soup = BeautifulSoup(requests.get(link).content, "html.parser")
    
    id = link.split("/")[4]
    
    anime_dict = {}
    anime_dict['Recommendations'] = "-"
    anime_dict['Title'] = "-"
    anime_dict['Ranked'] = "-"
    anime_dict['Popularity'] = "-"
    anime_dict['Members'] = ""
    anime_dict['PositiveReviews'] = ""
    anime_dict['MixedReviews'] = ""
    anime_dict['NegativeReviews'] = ""
    anime_dict['Staff'] = ""
    anime_dict['Type'] = ""
    anime_dict['Episodes'] = ""
    anime_dict['Status'] = ""
    anime_dict['Aired'] = ""
    anime_dict['Producers'] = ""
    anime_dict['Licensors'] = ""
    anime_dict['Studios'] = ""
    anime_dict['Source'] = ""
    anime_dict['Genres'] = ""
    anime_dict['Themes'] = ""
    anime_dict['Demographic'] = ""
    anime_dict['Duration'] = ""
    anime_dict['Rating'] = ""
    anime_dict['Related_content'] = ""

    rec_dict = {}
    # show recomendations (bottom of the page)
    try:
        for i in show_soup.findAll("li", {"class": "btn-anime", "title": True}):
            show_title = i['title']
            users_rec = i.find("span", class_="users").text.split(" ")[0]
            if(users_rec == "AutoRec"):
                break 
            rec_dict[show_title] = users_rec
    except Exception as e:
        print(id, "Error scraping recommendations", e)

    anime_dict["Recommendations"] = rec_dict

    # easy-scrappable data
    things = show_soup.findAll("strong")
    try:
        anime_dict['Title'] = things[0].text
    except Exception as e:
        print(id, "Error scraping title", e)
    try:
        anime_dict['Ranked'] = things[1].text
    except Exception as e:
        print(id, e)
    try:
        anime_dict['Popularity'] = things[2].text
    except Exception as e:
        print(id, e)
    try: 
        anime_dict['Members'] = things[3].text
    except Exception as e:
        print(id, e)
    try:
        anime_dict['PositiveReviews'] = things[4].text
    except Exception as e:
        print(id, e)
    try:
        anime_dict['MixedReviews'] = things[5].text
    except Exception as e:
        print(id, e)
    try:
        anime_dict['NegativeReviews'] = things[6].text
    except Exception as e:
        print(id, "Error scraping reviews", e)

    # staff
    staff_dict = []
    try:
        for i in show_soup.findAll("div", class_="detail-characters-list")[1].findAll("td", attrs={"valign":"top"}):
            charge = i.find("a", href=True).text.strip()
            name = ""
            if i.find("small") != None:
                name = i.find("small").text.split(",")[0]
                staff_dict.append([charge, name])
    except Exception as e:
        print(id, "Error scraping staff", e)

    anime_dict['Staff'] = staff_dict

    # misc. info
    index = 0
    
    x = show_soup.findAll("div", "spaceit_pad")

    initial_idx = 3

    lans = ['German', 'Spanish', 'French', 'English']
    for idx, i in enumerate(x[3:]):
        lans_count = 0
        for lan in lans:
            if lan not in i.text.strip():
                lans_count += 1
        if lans_count == 4:
            initial_idx += idx
            break

    try:
        anime_dict['Type'] = x[initial_idx].text.strip().split(":")
        anime_dict['Episodes'] = x[initial_idx + 1].text.strip().split(":")
        anime_dict['Status'] = x[initial_idx + 2].text.strip().split(":")
        anime_dict['Aired'] = x[initial_idx + 3].text.strip().split(":")
    except Exception as e:
        print(id, "Error on type until aired", e)

    try:
        anime_dict['Producers'] = x[initial_idx + 6].text.strip().split(":")
        anime_dict['Licensors'] = x[initial_idx + 7].text.strip().split(":")
        anime_dict['Studios'] = x[initial_idx + 8].text.strip().split(":")
        anime_dict['Source'] = x[initial_idx + 9].text.strip().split(":")
    except Exception as e:
        print(id, "Error on producers until source", e)
    try:
        anime_dict['Genres'] = x[initial_idx + 10].text.strip().split(":")
        anime_dict['Themes'] = x[initial_idx + 11].text.strip().split(":")
    except Exception as e:
        print(id, "Error on genres and themes", e)
    try:
        anime_dict['Demographic'] = x[initial_idx + 12].text.strip().split(":")
        anime_dict['Duration'] = x[initial_idx + 13].text.strip().split(":")
        anime_dict['Rating'] = x[initial_idx + 14].text.strip().split(":")
    except Exception as e:
        print(id, "Error on demographic and duration" ,e)

    # related content    
    related_dict = {}
    try:
        for i in show_soup.find("table", class_="anime_detail_related_anime").findAll("tr"):
            related_dict[i.find("td").text] = i.find("a", href=True)['href']
    except Exception as e:
        print(id, "Error scraping Related Content", e)

    anime_dict['Related_content'] = related_dict

    return id, anime_dict
````

````python
idx, res = scrap_all_raw("https://myanimelist.net/anime/40591/Kaguya-sama_wa_Kokurasetai_Tensai-tachi_no_Renai_Zunousen")
````

````python
data = res

def flatten_data_row(data, idx):
    flattened_data = {}
    recommendations = data['Recommendations']
    flattened_data['index'] = idx
    for i, (anime, rating) in enumerate(recommendations.items(), start=1):
        if i > 10:
            break
        flattened_data[f'rec_{i}'] = anime

    # Pad recommendations columns with NaN if less than 10
    for i in range(len(flattened_data) + 1, 11):
        flattened_data[f'rec_{i}'] = None

    # Staff
    staff = data['Staff']
    for i, s in enumerate(staff, start=1):
        flattened_data[f'staff_{i}'] = ':'.join(s)

    # Pad staff columns with NaN if less than 4
    for i in range(len(staff) + 1, 5):
        flattened_data[f'staff_{i}'] = None

    for key, value in data.items():
        if isinstance(value, list):
            flattened_data[key] = ', '.join(map(str, value))  # Convert all elements to strings before joining
        elif isinstance(value, dict):
            flattened_data[key] = ', '.join([f'{k}: {v}' for k, v in value.items()])
        else:
            flattened_data[key] = value
    del flattened_data['Recommendations']
    print(flattened_data)
    return flattened_data
````

````python
def flatten_dataframe(data):
    df_in_list = []
    for key in data:
        df_in_list.append(flatten_data_row(data[key], key))

    df = pd.DataFrame(df_in_list)
    return df
````

````python
base_url = "https://myanimelist.net/anime/season/"
years = ["2022"]
seasons = ["/fall"]
````

````python
for year in years:
    for season in seasons:
        page_url = base_url + year + season
        page = requests.get(page_url).content
        soup = BeautifulSoup(page, 'html.parser')
        print(year, season[1:])
        df = scrap_season_soup(soup, years[0], seasons[0])
        df.to_csv(f"df_{year}_{season[1:]}.csv", index=False)
        time.sleep(300)
````

````python
def scrap_season_soup(soup, year, season):
    # category index used to avoid TV continuing shows
    cat_index = 0
    season_scrapped_data = {}
    for x in soup.findAll("div", class_="seasonal-anime-list"):
        # print("\n====|", x.find("div").text, "|====") # i.e. ====| TV (New) |====
        index = 0
        if cat_index != 1:
            for show in x.findAll("div", class_="js-seasonal-anime"):
                #print(str(index) + ".",end=" ") # just to do 43. url \n 44. url \n 45. url
                show_url = show.find("a", class_="link-title")['href']

                try:
                    idx, result = scrap_all_raw(show_url)
                    season_scrapped_data[idx] = result
                except Exception as e:
                    print("Got an error on show with index:", show_url.split("/")[4])
                    print(e)

                #print(show_url)
                #index += 1
        cat_index += 1
    return flatten_dataframe(season_scrapped_data)
````
