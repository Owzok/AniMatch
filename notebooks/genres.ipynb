```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("data_anime_almclean.csv", index_col="Id")["Genres"]
df = pd.DataFrame(df)
df.head(30)
```

```python
def clean_genres(genres):
    genre_list = genres.split(', ')
    cleaned_genres = []
    if len(genre_list) == 1:
        genre_list[0] = genre_list[0]
    for genre in genre_list:
        m_len = int(len(genre)/2)
        if genre[:m_len] == genre[m_len:]:
            cleaned_genres.append(genre[:m_len])
        else:
            parts = genre.split()
            if len(parts) == 2 and parts[0] == parts[1]:
                cleaned_genres.append(parts[0])
            else:
                cleaned_genres.append(genre)

    return ', '.join(cleaned_genres)

df['Genres'] = df['Genres'].apply(clean_genres)
df.head(30)
```

```python
df.loc[9260]
```

```python
df['Genres'] = df['Genres'].str.split(', ')
df
```

```python
all_genres = set(genre for genres_list in df['Genres'] for genre in genres_list)
for genre in all_genres:
    df[genre] = 0

df
```

```python

for index, row in df.iterrows():
    genres_list = row['Genres']
    for genre in genres_list:
        df.at[index, genre] = 1

df.drop(columns=['Genres'], inplace=True)

print(df)
```

```python
df
```

```python
def find_similar_ids(df, target_id, num_similar=20):
    target_genres = df.loc[target_id, :].values.reshape(1, -1)
    similarities = cosine_similarity(target_genres, df.values)
    similar_indices = similarities.argsort()[0][::-1][1:num_similar + 1]
    similar_ids = df.index[similar_indices].tolist()

    return similar_ids
```
