```python
df = pd.DataFrame(df['Studios'])
df
```

```python
df.iloc[1]['Studios'] = 'Sunrise,Bones'
df.head(5)
```

```python
df['Studios'] = df['Studios'].apply(lambda x: x.split(","))
df.head(5)
```

```python
all_studios = set(studio for studio_list in df['Studios'] for studio in studio_list)
for studio in all_studios:
    df[studio] = 0

df
```

```python

for index, row in df.iterrows():
    studio_list = row['Studios']
    for studio in studio_list:
        df.at[index, studio] = 1

df.drop(columns=['Studios'], inplace=True)

print(df)
```

```python
df.to_csv("Studios.csv",index=True)
```
